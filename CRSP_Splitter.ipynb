```python
import shutil
import pandas as pd
import glob
import os
```

```python
filters = pd.read_csv("filters.csv")
filters_join = filters[(filters.Ticker.str.contains(";")) & (filters.Ivol_has_it == "1")].Ticker.to_list()
```

```python
for i in range(len(filters_join)):
    x = filters_join[i].split(" ; ")
    print(x)
    old_crsp_path = glob.glob(f"{os.getcwd()}\\crsp_ind\\{x[0]}*.csv")[0]
    new_crsp_path = glob.glob(f"{os.getcwd()}\\crsp_ind\\{x[1]}*.csv")[0]
    old_crsp = pd.read_csv(old_crsp_path)
    new_crsp = pd.read_csv(new_crsp_path)
    shutil.move(old_crsp_path,old_crsp_path.replace("crsp_ind","crsp_move"))
    shutil.move(new_crsp_path,new_crsp_path.replace("crsp_ind","crsp_move"))
    final_df = pd.concat([old_crsp,new_crsp])
    final_df.to_csv(f"{os.getcwd()}\\crsp_ind\\{x[0]}_ind.csv")
    print(x, "crsp done")
    old_ivol_path = glob.glob(f"{os.getcwd()}\\ind_1\\{x[0]}_ind.csv")[0]
    new_ivol_path = glob.glob(f"{os.getcwd()}\\ind_1\\{x[1]}_ind.csv")[0]
    old_ivol = pd.read_csv(old_ivol_path)
    new_ivol = pd.read_csv(new_ivol_path)
    shutil.move(old_ivol_path,old_ivol_path.replace("ind_1","ind_move"))
    shutil.move(new_ivol_path,new_ivol_path.replace("ind_1","ind_move"))
    final_df = pd.concat([old_ivol,new_ivol])
    final_df.to_csv(f"{os.getcwd()}\\ind_1\\{x[0]}_ind.csv")
    print(x, "ivol done")
```
